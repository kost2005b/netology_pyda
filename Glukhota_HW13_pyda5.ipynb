Мы хотим найти прямую, которая разделит точки в пространстве наполовину. Это можно сделать графически. Нарисовать график. Можно ли сказать, что данная прямая является собственным вектором? При проецировании данных на эту прямую разброс данных максимален?

```python
ladder = data['Ladder']
generosity = data['Generosity']
sns.regplot(x= ladder, y=generosity,data=data, fit_reg=True)
plt.xlabel('Удовлетворение жизнью')
plt.ylabel('Доброта')
plt.title('Корреляционная зависимость двух выборок')
plt.grid(False)
plt.show()
```

Этот график в виде прямой показывает вектор взаимосвязи двух выборок, расстояние от точек выборки до прямой будет стремиться к минимальному. Собственный же вектор, напротив, строится так, что расстояние до него у точек выборки получается максимальным, т.к. показывает размах

# Дискретное распределение случайной величины

Напишите код для моделирования 10 подбрасываний игрального кубика. Начертите диаграму histogram. Какое это распределение?

```python
#10 подбрасываний - биномиальное распределение
cube = stats.binom.rvs(n=10,p=0.166,size=50)

plt.hist(cube, bins = 6)
plt.xlabel('Количество выпадений стороны кубика в 10 подбрасываниях')
plt.ylabel('Встречаемость грани')
plt.show()
```

Напишите код для моделирования 1000 подбрасываний игрального кубика. Начертите диаграму histogram. Почему график изменился? Как применить центральную предельную теорему?

```python
#1000 подбрасываний - распределение стремится к нормальному.
cube_upd = stats.binom.rvs(n=1000,p=0.166,size=50)

plt.hist(cube_upd, bins = 6)
plt.xlabel('Количество выпадений стороны кубика в 1000 подбрасываниях')
plt.ylabel('Встречаемость грани')
plt.show()
```

 ЦПТ применяется в том, что увеличение числа экспериментов приближает распределение к нормальному. Каждый отдельный бросок дает с равной вероятностью какую-то одну грань, но из-за того, что каждый бросок не влияет на следующий, это - независимые эксперименты и изначально их распределение биномиально. На 10 бросках мы гарантированно получаем искажение - в самом равномерном исходе 4 грани выпадут дважды, а скорее всего как минимум одна грань - трижды. Увеличение числа экспериментов нивелируте эффект того, что равновероятных значений у нас 6
