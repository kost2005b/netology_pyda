### Задание 3:

Используя numpy, посчитайте значение 3-ей и 4-ой производных функций в точке:

$f(x) = x^5+4sin(2x)+cos(3x+3)$ в точке $x=1$


```python
def f(x):
    return x ** 5 + 4 * np.sin(x * 2) + np.cos(x * 3 + 3)
```

```python
x = np.linspace(-10, 10)
fx = f(x)
f1x = [derivative(f, xx, dx=1e-6, n=3, order=5) for xx in x]
```

```python
plt.plot(fx, label='function')
plt.plot(f1x, label='derivative')
plt.legend()
plt.show()
```

```python
x_ = np.linspace(-10, 10)
fx_ = f(x_)
f1x_ = [derivative(f, xx, dx=1e-6, n=4, order=5) for xx in x_]
```

```python
plt.plot(fx, label='function')
plt.plot(f1x_, label='derivative')
plt.legend()
plt.show()
```
